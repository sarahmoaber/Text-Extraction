```python
# Import OpenCV library for image and video processing
import cv2

# Import easyocr library to recognize text in images
import easyocr
```

```python
# Upload the image from which you want to extract text.
from google.colab.patches import cv2_imshow

from google.colab import files
uploaded = files.upload()
```

```python
# Read the image from the specified file path
img = cv2.imread('/content/photo5.jpg')

# Display the image in the Google Colab environment
cv2_imshow(img)
```

```python
# Initialize the EasyOCR reader for English language
reader = easyocr.Reader(['en'])

# Use the reader to detect text in the provided image
result = reader.readtext(img)

# Loop through each detected text in the result
for text in result:
    # Extract the top-left coordinate of the bounding box
    p1 = text[0][0]

    # Extract the bottom-right coordinate of the bounding box
    p2 = text[0][2]

    # Extract the detected text itself
    detectedText = text[1]

    # Draw a rectangle around the detected text on the image using the coordinates
    cv2.rectangle(img, p1, p2, (0, 255, 0), 3)

    # Overlay the detected text on the image near the bounding box
    cv2.putText(img, detectedText, p1, cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

# Display the image with the detected text and bounding boxes
cv2_imshow(img)
```
